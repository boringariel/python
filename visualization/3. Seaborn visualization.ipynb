## Seaborn visualization
---

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./target.csv', index_col=0).T
df2 = df.T
```

```python
# mRNA, miRNA microarray data로 correlation test를 수행한 결과를 이용한다
df.head()
```

---
### Heatmap

```python
plt.figure(figsize=(23,9))
sns.heatmap(df, cmap="YlGnBu")
plt.title('correlation analysis', fontsize=20)

plt.show() 
```

---
### Log-scaling Heatmap

```python
raw = pd.read_csv('./raw.csv', index_col = 0)
```

```python
from matplotlib.colors import LogNorm
```

```python
plt.figure(figsize=(8,13))
sns.heatmap(raw, cmap="Blues", annot = True, norm = LogNorm(vmin = raw['multifocal'].min(), vmax = raw['multifocal'].max()))
plt.title('miRNA expression', fontsize=20)

plt.show()
```

---
### Boxplot

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,10))
sns.boxplot(data = df2)
```

---
### Barplot

```python
sns.set(font_scale = 1.2)         #seaborn style 지정. 글꼴이나 폰트 크기 등 설정 가능
ax = raw.plot(kind = 'bar', figsize = (15,10), logy = True)    #logy: log-scaling
plt.title('miRNA expression', fontsize=20)
ax.set(ylabel = 'expression(log)')
```
