## Seaborn을 이용해 데이터 시각화하기
---

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./target.csv', index_col=0).T
df2 = df.T
```

```python
# mRNA, miRNA microarray data로 correlation test를 수행한 결과를 이용한다
df.head()
```

---
### heatmap 그리기

```python
plt.figure(figsize=(23,9))
sns.heatmap(df, cmap="YlGnBu")
plt.title('correlation analysis', fontsize=20)

plt.show() 
```

---
### Boxplot 그리기

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,10))
sns.boxplot(data = df2)
```
