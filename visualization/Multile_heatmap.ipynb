## Multiple heatmap
---

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

#seaborn Rd-Bl-Gn palette
pal = sb.diverging_palette(12.2, 127.7, l=53.4, n=255, center="dark")
```

```python
df = pd.read_csv("./gse65858_2.csv", index_col=0)
```

```python
df.head()
```

```python
sb.set(font_scale=1.1)
```

```python
df1 = df.iloc[:,0:73]
df2 = df.iloc[:,73:-1]

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (20, 12))
fig.subplots_adjust(wspace=0.03)
    
sb.heatmap(df1, cmap=pal, ax=ax1, cbar=False)
fig.colorbar(ax1.collections[0], ax=ax1, location="left", use_gridspec=False, pad=0.15)
 
sb.heatmap(df2, cmap=pal, ax=ax2, cbar=False)
 
ax2.yaxis.tick_right()

plt.text(-100, -0.2, r'positive', fontsize=24)
plt.text(80, -0.2, r'negative', fontsize=24)

plt.savefig('./res.png')
```

```python
plt.subplots(figsize = (20, 12))
ax = sb.heatmap(df, cmap = pal)
plt.text(25, -0.2, r'positive', fontsize=24)
plt.text(150, -0.2, r'negative', fontsize=24)

ax.vlines ([73], * ax.get_ylim (), colors = 'white')
```
