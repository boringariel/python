## histogram, scatterplot 만들기
* iris data를 이용한 다양한 시각화 구현하기
* matplotlib를 이용함
---

```python
import matplotlib
matplotlib.style.use('ggplot') #ggplot 테마를 사용함
import numpy as np #실수 범위의 np.arange(시작,끝,단위) 사용 가능
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f=open('./iris.txt')
lines=f.readlines() #각 행별로 리스트화해서 읽어들임
```

```python
lines[0:10]
```

```python
sepal_length_setosa=[]
sepal_length_versi=[]
sepal_length_virgi=[]
```

```python
print(lines[0]) #각 정보는 탭(\t)으로 구분됨.마지막은 계행문자가 있음
```

```python
for line in lines[1:]:
    cells=line.strip().split('\t')
    print(cells)
```

```python
for line in lines[1:]:
    cells=line.strip().split('\t') #strip(): 양 끝의 계행문자를 없애줌
    if cells[4]=='I.setosa':
        sepal_length_setosa.append(float(cells[0]))
    elif cells[4]=='I.versicolor':
        sepal_length_versi.append(float(cells[0]))
    else:
        sepal_length_virgi.append(float(cells[0]))
```

---
### histogram 만들기
---

```python
plt.hist(sepal_length_setosa,bins=range(4,10)) #히스토그램 작성. bins는 계급구간 작성시 사용
plt.hist(sepal_length_versi,bins=range(4,10))
plt.hist(sepal_length_virgi,bins=range(4,10))
plt.show()
```

```python
plt.hist(sepal_length_setosa,bins=np.arange(4,10,0.1),label='I.setosa') #히스토그램 작성
plt.hist(sepal_length_versi,bins=np.arange(4,10,0.1),label='I.versicolor')
plt.hist(sepal_length_virgi,bins=np.arange(4,10,0.1),label='I.verginica')
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('count')
plt.show()
```

```python
sepal_width_setosa=[]
sepal_width_versi=[]
sepal_width_virgi=[]
```

```python
for line in lines[1:]:
    cells=line.strip().split('\t')
    if cells[4]=='I.setosa':
        sepal_width_setosa.append(float(cells[1]))
    elif cells[4]=='I.versicolor':
        sepal_width_versi.append(float(cells[1]))
    else:
        sepal_width_virgi.append(float(cells[1]))
```

```python
plt.hist(sepal_width_setosa,bins=np.arange(0,5,0.1),label='I.setosa')
plt.hist(sepal_width_versi,bins=np.arange(0,5,0.1),label='I.versicolor')
plt.hist(sepal_width_virgi,bins=np.arange(0,5,0.1),label='I.verginica')
plt.legend()
plt.xlabel('sepal width (cm)')
plt.ylabel('count')
plt.show()
```

---
### scatterplot 만들기
---

```python
plt.scatter(sepal_length_setosa,sepal_width_setosa,label='I.setosa') #scattler(x값, y값) 지정해 분포표 작성 가능
plt.scatter(sepal_length_versi,sepal_width_versi,label='I.versicolor')
plt.scatter(sepal_length_virgi,sepal_width_virgi,label='I.virginica')
plt.legend()
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Comparison among species')
plt.show()
```

---
### 각 분류별로 histogram 만들기
---

```python
sepal_length_setosa=[]
sepal_length_versi=[]
sepal_length_virgi=[]

sepal_width_setosa=[]
sepal_width_versi=[]
sepal_width_virgi=[]

petal_length_setosa=[]
petal_length_versi=[]
petal_length_virgi=[]

petal_width_setosa=[]
petal_width_versi=[]
petal_width_virgi=[]
```

```python
for line in lines[1:]:
    cells=line.strip().split('\t')
    if cells[4]=='I.setosa':
        sepal_length_setosa.append(float(cells[0]))
        sepal_width_setosa.append(float(cells[1]))
        petal_length_setosa.append(float(cells[2]))
        petal_width_setosa.append(float(cells[3]))
    elif cells[4]=='I.versicolor':
        sepal_length_versi.append(float(cells[0]))
        sepal_width_versi.append(float(cells[1]))
        petal_length_versi.append(float(cells[2]))
        petal_width_versi.append(float(cells[3]))
    else:
        sepal_length_virgi.append(float(cells[0]))
        sepal_width_virgi.append(float(cells[1]))
        petal_length_virgi.append(float(cells[2]))
        petal_width_virgi.append(float(cells[3]))
```

```python
#plt.figure(figsize=(10,5)) #도표 크기 지정

fig,ax=plt.subplots(4,figsize=(7,10)) #4개 도표 작성

ax[0].hist(sepal_length_setosa,bins=np.arange(0,10,0.1),label='I.setosa (sl)')
ax[0].hist(sepal_length_versi,bins=np.arange(0,10,0.1),label='I.versicolor (sl)')
ax[0].hist(sepal_length_virgi,bins=np.arange(0,10,0.1),label='I.verginica (sl)')
ax[0].set_xlabel('sepal length(cm)') #subplot 설정시에는 set_xlabel 꼴로
ax[0].set_ylabel('count')
ax[0].legend()

ax[1].hist(sepal_width_setosa,bins=np.arange(0,10,0.1),label='I.setosa (sw)')
ax[1].hist(sepal_width_versi,bins=np.arange(0,10,0.1),label='I.versicolor (sw)')
ax[1].hist(sepal_width_virgi,bins=np.arange(0,10,0.1),label='I.verginica (sw)')
ax[1].set_xlabel('sepal width(cm)')
ax[1].set_ylabel('count')
ax[1].legend()

ax[2].hist(petal_length_setosa,bins=np.arange(0,10,0.1),label='I.setosa (pl)')
ax[2].hist(petal_length_versi,bins=np.arange(0,10,0.1),label='I.versicolor (pl)')
ax[2].hist(petal_length_virgi,bins=np.arange(0,10,0.1),label='I.verginica (pl)')
ax[2].set_xlabel('petal length(cm)')
ax[2].set_ylabel('count')
ax[2].legend()

ax[3].hist(petal_width_setosa,bins=np.arange(0,10,0.1),label='I.setosa (pw)')
ax[3].hist(petal_width_versi,bins=np.arange(0,10,0.1),label='I.versicolor (pw)')
ax[3].hist(petal_width_virgi,bins=np.arange(0,10,0.1),label='I.verginica (pw)')
ax[3].set_xlabel('petal width(cm)')
ax[3].set_ylabel('count')
ax[3].legend()

#plt.legend()
#plt.xlabel('cm')
#plt.ylabel('count')
plt.tight_layout() #여백 확보    
plt.show()
```

### pandas를 이용해 그래프 그리기. 반복문을 사용하지 않음
---

```python
df_iris=pd.read_csv('./iris.txt',sep='\t')
```

```python
df_iris.head()
```

```python
m=(df_iris['Species']=='I.setosa') #데이터프레임의 Species 항목별로 입력값을 분리함
n=(df_iris['Species']=='I.versicolor')
o=(df_iris['Species']=='I.virginica')
```

```python
is_sl=df_iris[m]['Sepal length'] #입력값 중, 특정 조건을 만족하는대로 입력값을 분리함
ive_sl=df_iris[n]['Sepal length']
ivi_sl=df_iris[o]['Sepal length']

is_sw=df_iris[m]['Sepal width']
ive_sw=df_iris[n]['Sepal width']
ivi_sw=df_iris[o]['Sepal width']

is_pl=df_iris[m]['Petal length']
ive_pl=df_iris[n]['Petal length']
ivi_pl=df_iris[o]['Petal length']

is_pw=df_iris[m]['Petal width']
ive_pw=df_iris[n]['Petal width']
ivi_pw=df_iris[o]['Petal width']
```

```python
fig,bx=plt.subplots(4,figsize=(7,10))

bx[0].hist(is_sl,bins=np.arange(0,10,0.1),label='I.setosa')
bx[0].hist(ive_sl,bins=np.arange(0,10,0.1),label='I.versicolor')
bx[0].hist(ivi_sl,bins=np.arange(0,10,0.1),label='I.virginica')
bx[0].set_xlabel('sepal length(cm)')
bx[0].set_ylabel('count')
bx[0].legend()

bx[1].hist(is_sw,bins=np.arange(0,10,0.1),label='I.setosa')
bx[1].hist(ive_sw,bins=np.arange(0,10,0.1),label='I.versicolor')
bx[1].hist(ivi_sw,bins=np.arange(0,10,0.1),label='I.virginica')
bx[1].set_xlabel('sepal width(cm)')
bx[1].set_ylabel('count')
bx[1].legend()

bx[2].hist(is_pl,bins=np.arange(0,10,0.1),label='I.setosa')
bx[2].hist(ive_pl,bins=np.arange(0,10,0.1),label='I.versicolor')
bx[2].hist(ivi_pl,bins=np.arange(0,10,0.1),label='I.virginica')
bx[2].set_xlabel('petal length(cm)')
bx[2].set_ylabel('count')
bx[2].legend()

bx[3].hist(is_pw,bins=np.arange(0,10,0.1),label='I.setosa')
bx[3].hist(ive_pw,bins=np.arange(0,10,0.1),label='I.versicolor')
bx[3].hist(ivi_pw,bins=np.arange(0,10,0.1),label='I.virginica')
bx[3].set_xlabel('petal width(cm)')
bx[3].set_ylabel('count')
bx[3].legend()

plt.tight_layout()
plt.show()
```

```python
df=df_iris #데이터프레임 값을 변수로 선언

fig,ax=plt.subplots(2,figsize=(5,7)) #분포표 작성 (2개)

df[m].plot.scatter('Sepal length','Sepal width',c='b',label='I.setosa',ax=ax[0],legend=False)
df[n].plot.scatter('Sepal length','Sepal width',c='r',label='I.versicolor',ax=ax[0],legend=False)
df[o].plot.scatter('Sepal length','Sepal width',c='y',label='I.virginica',ax=ax[0],legend=False)
#지정한 요소별로 마스킹. 색상 지정. label 지정. 표 번호 지정. 범례 미표시

df[m].plot.scatter('Petal length','Petal width',c='b',label='I.setosa',ax=ax[1])
df[n].plot.scatter('Petal length','Petal width',c='r',label='I.versicolor',ax=ax[1])
df[o].plot.scatter('Petal length','Petal width',c='y',label='I.virginica',ax=ax[1])

plt.show()
```
