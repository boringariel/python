## 이미지 불러오기
* 파이썬에서는 3차원 array 형태로 이미지를 처리할 수 있으며, 기계학습시 해당 처리법을 주로 이용한다
* 해당 이미지 데이터는 plot으로 그려 시각화할 수 있다
---

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_sample_images
```

```python
d = load_sample_images()
```

```python
#파이썬의 이미지 구조. 세로/가로/색상 정보로 구성됨
d.images[0].shape
```

```python
#plot 형식으로 두 장의 이미지 출력하기
fig, axes = plt.subplots(1, 2, figsize = (10,10))
ax1 = axes[0]
ax2 = axes[1]
ax1.imshow(d.images[0])
ax1.axis('off') #축 없애기
ax2.imshow(d.images[1])
plt.show()
```

```python
#위와 같은 출력을 반복문을 활용해 수행한다
fig, axes = plt.subplots(1, 2, figsize = (10,10))
for n,ax in enumerate(np.ravel(axes)): #enumerate: 번호를 할당해줌
    ax.imshow(d.images[n])
    ax.axis('off')
plt.show()
```
